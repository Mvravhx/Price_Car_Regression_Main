```python
import numpy as np
import pandas as pd
# Importo OS solo para ver la lista de archivos de data
import os
```

```python
lista_archivos = os.listdir("./data")

df = pd.read_csv(f"./data/{lista_archivos[0]}")

for archivo in lista_archivos[1:]:
    
    if "km_77" in archivo:
    
        df2= pd.read_csv(f"./data/{archivo}")

        df = pd.concat([df, df2], axis = 0)

```

```python
df = df.drop_duplicates(keep= "first")
df.reset_index(inplace= True, drop=True)
```

```python
valores = []

for i in df["Potencia"].values:
    try:

        valores.append(int(i.split(" ")[0]))

    except:

        valores.append(0)

df["Potencia"] = valores
```

```python
valores = []

for i in df["Velocidad máxima"].values:
    try:

        valores.append(float(i.split(" ")[0]))

    except:

        valores.append(0)

df["Velocidad máxima"] = valores
        
```

```python
valores = []

for i in df["Aceleracion"].values:
    
    try:

        valores.append(float(i.split(" ")[0]))

    except:

        valores.append(0)

df["Aceleracion"] = valores
```

```python
valores = []

for i in df["Consumo"].values:
    
    try:

        valores.append(float(i.split(" ")[0]))

    except:

        valores.append(0)

df["Consumo"] = valores
        
```

```python
valores = []

for i in df["Emisiones"].values:
    
    try:

        valores.append(float(i.split("g ")[0]))

    except:

        valores.append(-1)

df["Emisiones"] = valores
        
```

```python
valores = []

for i in df["KM"].values:
    
    if type(i) == str:
        
        valores.append(int(i.replace(".","")))

        continue

    valores.append(int(i * 1000))

df["KM"] = valores
```

```python
df = df[~df['Combustible'].isna()]
combustibles_a_eliminar = ['?', 'Sin tracción', 'Super']

df = df[~df['Combustible'].isin(combustibles_a_eliminar)]
df['Combustible'].replace(['GASOLINA', 'gasolina'], 'Gasolina', inplace=True)
df['Combustible'].replace(['ELÉCTRICO'], 'Eléctrico', inplace=True)
df['Combustible'].replace(['Híbrido (Gasolina)', 'Híbrido Enchufable (PHEV)', 'Híbrido (Diésel)'], 'Híbrido', inplace=True)
df['Combustible'].replace(['Gas Licuado (GLP)', 'Gas Natural (GNC)', 'GNC', 'Etanol'], 'Gas', inplace=True)
df['Combustible'].replace(['Gas', 'Hidrógeno'], 'Otros', inplace=True)
# df = df.drop(df[df['Combustible'] == 'Otros'].index)
```

```python
df[['Largo', 'Ancho', 'Alto']] = df['Dimensiones'].str.split('mm /', expand=True)
df['Alto'] = df['Alto'].str.replace(' mm', '')
df['Largo'] = df['Largo'].str.replace(' ','')
```

```python
columns = ["Largo","Ancho","Alto"]

for column in columns:
    valores = [] 
    for i in df[column].values:
        try:
            valores.append(int(i))
        except:
            valores.append(0)

    df[column] = valores
```

```python
#Rellenamos los nans con la moda, en este caso Toledo
df['Ciudad'] = df['Ciudad'].fillna(df['Ciudad'].mode()[0])
```

```python
df["Precio"] = df["Precio"].apply(lambda x : int(x[:-1].replace(".","")))

df["Etiqueta medioambiental"] = df["Etiqueta medioambiental"].apply(lambda x : "Sin etiqueta" if x == "False" else x)

df["Nº puertas"] = df["Nº puertas"].apply(lambda x : "0" if x == "False" else x)
```

```python
df[["Potencia","Año","Velocidad máxima","Emisiones","Largo","Ancho","Alto"]] = df[["Potencia","Año","Velocidad máxima","Emisiones","Largo","Ancho","Alto"]].astype("int16")
df[["KM"]] = df[["KM"]].astype("int32")
df[["Nº puertas"]] = df[["Nº puertas"]].astype("int8")

df[["Aceleracion","Consumo"]] = df[["Aceleracion","Consumo"]].astype("float32")

df[["Combustible","Cambio","Tipo de carrocería","Etiqueta medioambiental"]] = df[["Combustible","Cambio","Tipo de carrocería","Etiqueta medioambiental"]].astype("category")

df["Garantía"] = df["Garantía"].astype("bool")
```

```python
# Creo diccionario para unificar colores
dic_colores = {
    'Rojo': 'rojo|granate|red|burdeos|rouge|bordeaux|rosa',
    'Azul': 'azul|blue|navy|bleu|turquoise|blau|azual|ocean|cerulea|still night|soldadita',
    'Blanco': 'bl|blanco|white|snow|pearl|nacar|nacarado|glacier|blanc|balnco|bianco|weiss|weiß|bkanco',
    'Gris': 'gris|plateado|platino|platinium|platinum|grey|gray|silver|penta metal|gr|plata|onyx|steel|plt|plat|silber|grau|gri|moonscape|stone|iron|grafito|shadow|gun',
    'Amarillo': 'amarillo|yellow|gold|lemon|dorado|oro|amarilla|brass',
    'Naranja': 'naranja|orange|mostaza|curcuma',
    'Negro': 'negro|black|negra|dark knight|obsidian|oscuro|nachtschwarz',
    'Beige': 'beige|champagne|champán|champan|dust|bílá',
    'Verde': 'verde|green|teal|liquen|forest|yucca',
    'Marron': 'marron|marrón|brown|caoba|bronce|bronze|copper|oak|ceramica|brun',
    'Violeta': 'violeta|purpura|púpura|berenjena',
}

# Creo loop para unificar las variables de los colores
for color, variables in dic_colores.items():
    # Creo booleana para las variables de los colores
    colores = df['Color'].str.contains(variables, case=False, na=False)

    # Reemplazo las variables de los colores por el color único 
    df.loc[colores, 'Color'] = color

# Restantes colores no determinados clasificados como 'Otro'
df.loc[~df['Color'].isin(dic_colores.keys()), 'Color'] = 'Otro'

# Verifico cantidad de colores
colores = df['Color'].unique()
cant_colores = len(colores)
print(f'Hay {cant_colores} colores únicos en el DF')
print('------------')
print(colores)
```

```python
#Elimino las filas en las que la marca sea False ya que el resto de características también están vacías
df = df[df['Marca'] != 'False']
```
